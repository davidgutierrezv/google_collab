FUNCION QUE CREA UN GRAFO

```python
import matplotlib.pyplot as plt
import networkx as nx

# Crear un grafo dirigido
G = nx.DiGraph()

# Nodos (entidades)
nodos = [
    "Nodo 1: Ciclo de Vida del Proyecto",
    "Nodo 2: Gestin de Proyectos",
    "Nodo 3: Artefactos Clave",
    "Nodo 4: Triple Restriccin",
    "Nodo 5: Stakeholders",
    "Nodo 6: Gestin de Riesgos",
    "Nodo 7: Control de Calidad",
    "Nodo 8: Evaluacin y Lecciones Aprendidas"
]

# Agregar nodos al grafo
G.add_nodes_from(nodos)

# Relaciones (aristas) entre nodos
relaciones = [
    ("Nodo 1: Ciclo de Vida del Proyecto", "Nodo 2: Gestin de Proyectos"),
    ("Nodo 1: Ciclo de Vida del Proyecto", "Nodo 3: Artefactos Clave"),
    ("Nodo 1: Ciclo de Vida del Proyecto", "Nodo 4: Triple Restriccin"),
    ("Nodo 2: Gestin de Proyectos", "Nodo 3: Artefactos Clave"),
    ("Nodo 2: Gestin de Proyectos", "Nodo 5: Stakeholders"),
    ("Nodo 3: Artefactos Clave", "Nodo 5: Stakeholders"),
    ("Nodo 3: Artefactos Clave", "Nodo 6: Gestin de Riesgos"),
    ("Nodo 3: Artefactos Clave", "Nodo 7: Control de Calidad"),
    ("Nodo 4: Triple Restriccin", "Nodo 8: Evaluacin y Lecciones Aprendidas"),
    ("Nodo 5: Stakeholders", "Nodo 7: Control de Calidad"),
    ("Nodo 6: Gestin de Riesgos", "Nodo 8: Evaluacin y Lecciones Aprendidas"),
    ("Nodo 7: Control de Calidad", "Nodo 8: Evaluacin y Lecciones Aprendidas")
]

# Agregar relaciones al grafo
G.add_edges_from(relaciones)

# Dibujar el grafo
plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G, seed=42)  # Posicionamiento de los nodos

# Dibujar los nodos y las aristas
nx.draw(G, pos, with_labels=True, node_size=3000, node_color="lightblue", font_size=10, font_weight="bold", edge_color="gray")

# Mostrar el grfico
plt.title("Modelo Relacional del Ciclo de Vida del Proyecto")
plt.show()
```
